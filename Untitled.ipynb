```python
import requests as rq
from bs4 import BeautifulSoup as bs
```

```python
url = "https://www.goodreads.com/book/show/6953508-some-we-love-some-we-hate-some-we-eat"
```

```python
page = rq.get(url)
page.status_code
```

```python
soup = bs(page.text, 'html')
print(soup)
```

```python
# Book Name
soup.find('div', class_ = "BookPageTitleSection__title").h1.text
```

```python
# Author Name
soup.find('span', class_ = "ContributorLink__name").text
```

```python
# Ratings 
soup.find('div', class_ = "RatingStatistics__rating").text
```

```python
# Number of ratings
soup.find('span', class_ = "ratingsCount").text
```
